```python
import pandas as pd
import matplotlib.pyplot as plt

# Відкриття файлу з отриманими даними
df = pd.read_excel("mydata.xlsx")

# Отримання даних з DataFrame
x_coords = df.iloc[1,1:].tolist()
print("X:", x_coords)
y_coords = df.iloc[2,1:].tolist()
print("Y:",y_coords)
radius = df.iloc[3,1:].tolist()
print("R",radius)
globmin = df.iloc[0,1]
print("Min radius: ",globmin)

# Середні координати для центру великого кола
x_center = (sum(x_coords) / len(x_coords)) -14
y_center = (sum(y_coords) / len(y_coords)) -2

fig, ax = plt.subplots()
# Додаємо велике коло
large_circle = plt.Circle((x_center, y_center), globmin, color='blue', fill=False, linewidth=2, label='Large Circle')
ax.add_patch(large_circle)

# Додаємо вписані кола
for x, y, r in zip(x_coords, y_coords, radius):
    circle = plt.Circle((x, y), r, color='red', fill=False, linewidth=2)
    ax.add_patch(circle)

ax.set_xlim(min(x_coords) - globmin, max(x_coords) + globmin)
ax.set_ylim(min(y_coords) - globmin, max(y_coords) + globmin)
ax.set_aspect('equal', adjustable='datalim')
plt.xlabel('X coordinate')
plt.ylabel('Y coordinate')
plt.legend()
plt.grid(True)
plt.title('Layout of Circles')
plt.show()
```
